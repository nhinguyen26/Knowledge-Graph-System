```python
mappings = dict()
```

```python
import csv

iter = 0

# opening the mapping found online, which is a mapping of entity ids to entity names
with open("mid2name.tsv") as tsv:
    for line in csv.reader(tsv, dialect="excel-tab"): 
      new_entity = line[0].lstrip('/').replace('/', '.')
      mappings[new_entity] = line[1]
```

```python
with open('train.txt') as f:
    train_lines = f.readlines()
```

```python
with open('test.txt') as f:
    test_lines = f.readlines()
```

```python
def write_mid_lines(text, file_name):
  for raw_line in text:
    line = raw_line.rstrip()
    split_line = line.split('\t')
    # converting id to name if the mapping exists
    if split_line[0] in mappings and split_line[-1] in mappings:
      with open(file_name, 'a') as file:
        file.write(mappings[split_line[0]] + '\t' + split_line[1] + '\t' + mappings[split_line[-1]] + '\n')
```

```python
write_mid_lines(test_lines, 'test_all_natural_language.txt')
write_mid_lines(train_lines, 'train_all_natural_language.txt')
```
